```python
#UNIVARIATE ANALYSIS
import pandas as pd
import seaborn as sns
data = pd.read_csv(r"C:\Users\ASUS\Downloads\vgsales.csv")
print(data.head(5))
```

```python
#Bargraph
sns.countplot(x=data['Genre'])
```

```python
#Bargraph
sns.countplot(x=data['Publisher'])
```

```python
# Pie chart to show the distribution of 'Publisher' in the dataset
import matplotlib.pyplot as plt
x = data['Publisher'].value_counts()
plt.pie(x.values, labels=x.index, autopct='%1.1f%%')
plt.show()
```

```python
# BIVARIATE ANALYSIS
# Categorical vs numerical
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 5))
sns.barplot(x=data['Genre'], y=data['Global_Sales'])
plt.xticks(rotation=90)
plt.show()
```

```python
# Numerical v/s Numerical
sns.scatterplot(x=data['NA_Sales'], y=data['EU_Sales'])
```

```python
# Categorical v/s Categorical
sns.countplot(x='Genre', hue='Platform', data=data)
```

```python
# MULTIVARIATE ANALYSIS
# principal component analysis
from sklearn import datasets, decomposition
from sklearn.preprocessing import LabelEncoder

# Drop rows with NaN values in the selected columns
data_cleaned = data.dropna(subset=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales'])

X = data_cleaned[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]
y = LabelEncoder().fit_transform(data_cleaned['Platform'])

pca = decomposition.PCA(n_components=2)
X = pca.fit_transform(X)
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
```

```python
# Heat map
import seaborn as sns
sns.heatmap(data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].corr(), annot=True)
```
